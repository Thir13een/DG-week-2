# Which company has maximum cab users at a particular time period?

```python
# Extract month and year for analysis
master['YearMonth'] = master['Date of Travel'].dt.to_period('M')
```

```python
master
```

```python
plot_ = master.groupby(['YearMonth', 'Company']).size().unstack(level='Company')
```

```python
plot_
```

```python
monthly_data = pd.DataFrame(plot_)
```

```python
# Reset the index to make 'YearMonth' a column (for easier plotting)
monthly_data.reset_index(inplace=True)
```

```python
monthly_data
```

```python
monthly_data.info()
```

```python
# Convert YearMonth to string for plotting purposes
monthly_data['YearMonth'] = monthly_data['YearMonth'].astype(str)

# Plot the data 
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_data, x='YearMonth', y='Pink Cab', marker='o', label='Pink Cab')
sns.lineplot(data=monthly_data, x='YearMonth', y='Yellow Cab', marker='o', label='Yellow Cab')

plt.title('Number of Users Over Time by Company')
plt.xlabel('Time (Year-Month)')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)  
plt.grid(True)

plt.legend(title='Company')

# Show the plot
plt.show()
```

**"Yellow Cab" is More Popular: "Yellow Cab" consistently has more users than "Pink Cab" throughout the entire time period.**

**The number of users for both "Yellow Cab" and "Pink Cab" gradually increases throughout the year, showing steady growth as the months progress.**

**Sharp Drop in December and January: There’s a noticeable sharp decline in user numbers during December and January. This drop could be due to holiday seasons or other factors that typically reduce cab usage.**

**Recovery After the Dip: After the dip in December and January, the number of users begins to rise again, continuing the pattern of growth as the year progresses.**

**BUT ALL IN ALL YELLOW CAB SERVICES ARE STILL FAVORED OVER PINK CAB SERVICES THROUGHOUT THE YEAR!!**

# What are the attributes of these customer segments?

```python
age_distribution = pd.DataFrame(master.groupby(['Company', 'Age']).size().unstack().T)

plt.figure(figsize=(12, 6))
sns.lineplot(data=age_distribution, palette = 'mako')

plt.title('Age Distribution of Users by Company')
plt.xlabel('Age')
plt.ylabel('Number of Users')
plt.grid(True)

# Show the legend to differentiate between the companies
plt.legend(title='Company')

# Show the plot
plt.show()
```

**"Yellow Cab" has a consistently higher number of users in the younger age groups (under 40) compared to "Pink Cab." This suggests that "Yellow Cab" is more popular among younger customers.**

**There's a noticeable sharp drop in the number of users for both "Pink Cab" and "Yellow Cab" around the age of 40. After the age of 40, the number of users stabilizes for both companies, with "Yellow Cab" maintaining a larger user base than "Pink Cab." However, the user count remains relatively low compared to younger age groups.**

**Across all age groups, "Yellow Cab" has more users than "Pink Cab," which matches what we've seen in other analyses where "Yellow Cab" consistently has more customers.**
